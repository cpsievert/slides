```julia
using RDatasets
ggdat = RDatasets.datasets("ggplot2")
```

```julia
using Gadfly
plot(dataset("ggplot2", "diamonds"), x = "Carat", y = "Price", Geom.hexbin)
```

```julia
using Interact
@manipulate for xn = 10:100, yn = 10:100
    plot(dataset("ggplot2", "diamonds"), x = "Carat", y = "Price", Geom.hexbin(xbincount = xn, ybincount = yn))
end
```
